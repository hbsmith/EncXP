It looks like there are 5 seed sets from the non-weighted version which have duplicate compounds in them. And the duplicate compound is always `C11505`

Let's look at the other seed sets

```python
all_seeds = dict()
for fname in glob.glob("../rseeds*.dat"):
    print(fname)
    with open(fname) as f:
        rows = csv.reader(f,delimiter=" ")
        csv_list = list(rows)
    
    [i.pop() for i in csv_list] # remove extra space 
    
    all_seeds[fname.split("/")[-1].split(".dat")[0]] = csv_list
    
    for i in csv_list:
        if len(set(i)) != len(i):
#             print(i,set(i))
#             print(len(set(i)))
            print(set([x for x in i if i.count(x) > 1]))
```

So we see that in both non-weighted sets, there are duplicates in some of the seeds.
And it's always `C11505`.

There are 5 duplicates in the 5nw.
There are 30 duplicates in the 10nw.

```python
for k in all_seeds:
    print(k,len([frozenset(i) for i in all_seeds[k]]),len(set([frozenset(i) for i in all_seeds[k]])))
```
